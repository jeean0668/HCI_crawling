```python
import pandas as pd
from urllib.request import urlopen
import requests as re
from urllib.parse import urlparse
from bs4 import BeautifulSoup
#from openpyxl import load_workbook
#import openpyxl
```

```python
# 파일 반드시 xls 혹은 csv로 바꿔줘야 합니다. 
df = pd.read_excel('input/NewsResult_20200301-20211231.xls', engine = 'xlrd')
df_row = list(df['URL'].dropna())
```

```python
articles = []
for link in df_row:
    print(link)
    if 'segye' in link:
        articles.append(link)
```

```python
with open("practice.txt", 'w', encoding='utf-8') as practice:
    for ar in articles:
        try:
            url = ar
            if not urlparse(url).scheme:
                url = 'http://'+url
            print(url)
            link_html = urlopen(url)
            link_bs = BeautifulSoup(link_html.read(), 'html.parser')
            for p_tags in link_bs.find('article', {'itemprop':'articleBody'}).findAll('p'):
                practice.write(p_tags.get_text())
        except:
            print("error ocurred")
```

```python

segye_articles = open("practice.txt",'r', encoding='utf-8')
data = []
for article in segye_articles:
    # 현재 article가 하나의 article별로 저장되므로, article 별로 다뤄줘야함.
    for line in article.splitlines():
        data.append(line)
# 개행문자 삽입
```

```python
from konlpy.tag import Hannanum, Kkma, Komoran, Okt

sample = '메밀 국수가 그리운 계절이 되었다.'

# 각각의 형태소 분석기를 초기화
han = Hannanum()
kkm = Kkma()
kom = Komoran()
okt = Okt()

# 분석하기 전의 문장 확인
print(sample)

print('------------------------------------')

# 각각의 형태소 분석기에서 형태소 분석을 한 결과를 출력, (형태소, 품사)의 형태로 출력됨
# 각 형태소 분석기가 품사 태그를 어떻게 정의했는지는 https://docs.google.com/spreadsheets/d/1OGAjUvalBuX-oZvZ_-9tEfYD2gQe7hTGsgUpiiBSXI8/edit#gid=0 참고
print(han.pos(sample))
print(kkm.pos(sample))
print(kom.pos(sample))
print(okt.pos(sample))

print('------------------------------------')

# 각각의 형태소 분석기에서 형태소 분석 결과에서, 품사 태그를 삭제하고 단어만 넣은 형태를 출력
print(han.morphs(sample))
print(kkm.morphs(sample))
print(kom.morphs(sample))
print(okt.morphs(sample))
from konlpy.tag import Hannanum, Kkma, Komoran, Okt
print('------------------------------------')

# 만약 품사 태그만을 추출하고 싶다면, 아래와 같은 간단한 for문으로 추출 가능
tags = []
for m in han.pos(sample):
  tags.append(m[1])
print(tags)
```

```python
# 앞에서 했던 빈도 분석 코드와 비슷한 내용을 준비
freq = defaultdict(int)

for line in data:
  # 단, 이번에는 각 문장을 형태소 분석하는 내용을 추가
  # 실습에서는 아주 간단한 처리만 수행하는 OKT 분석기를 사용함
  tokens = okt.morphs(line)
  for token in tokens:
    freq[token] += 1

# 위에서 했던 실습과 동일한 조건에서 결과를 비교
for f in freq:
  if freq[f] > 10:
    print(f, freq[f])
```

```python
import re
from collections import defaultdict
freq = defaultdict(int)
for line in data:
  # 이번 실습에서는 정규 표현식을 통해, 문장 내 단어의 패턴을 컴퓨터에게 인식시켜 처리하게 하는 방식을 사용
  # 가-힁 = 가에서 힇까지의 단어 (컴퓨터 내에 등록된 한글 단어들 중 맨 첫번째가 가, 맨 마지막이 힇)
  # ^<문자열> = <문자열>이 아닌 것
  # [<패턴>] = [] 안에 들어 있는 패턴에 해당하는 것
  # <패턴>+ = 해당 패턴의 글자가 1개나 그 이상이다
  # 종합해서, [^가-힁]+ = 한글 단어가 아닌 것이 1글자 이상 있으면
  # sub는 치환을 의미하는 명령어이기 때문에, 아래 코드를 글로 풀어서 설명하면,
  # line에 있는 데이터 중 한글 단어가 아닌 것이 1글자 이상 보이면 그것을 전부 없애버려라, 란 의미
  line = re.sub('[^가-힇]+', '', line) # 시간이 엄청걸림 
  # 형태소 분석은 동일하게 수행
  tokens = okt.morphs(line)
  for token in tokens:
    # len은 지정 변수의 길이를 재는 명령이어서, 이 경우 token의 길이가 1 이상일 경우에만 빈도를 추가
    if len(token) > 1: freq[token] +=1
    # freq[token] += 1

# 동일한 코드로 빈도 분석된 결과를 확인
for f in freq:
  if freq[f] > 10:
    print(f, freq[f])
```
