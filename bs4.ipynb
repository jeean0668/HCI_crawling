```python
import pandas as pd
from urllib.request import urlopen
import requests as re
from urllib.parse import urlparse
from bs4 import BeautifulSoup
#from openpyxl import load_workbook
#import openpyxl
```

```python
# 파일 반드시 xls로 바꿔줘야 합니다. 
df = pd.read_excel('input/NewsResult_20200301-20211231.xls', engine = 'xlrd')
df_row = list(df['URL'].dropna())
```

```python
articles = []
for link in df_row:
    print(link)
    if 'segye' in link:
        articles.append(link)
print(articles[:5])
```

```python
with open("practice.txt", 'w', encoding='utf-8') as practice:
    for ar in articles:
        try:
            url = ar
            if not urlparse(url).scheme:
                url = 'http://'+url
            print(url)
            link_html = urlopen(url)
            link_bs = BeautifulSoup(link_html.read(), 'html.parser')
            for p_tags in link_bs.find('article', {'itemprop':'articleBody'}).findAll('p'):
                practice.write(p_tags.get_text())
        except:
            print("error ocurred")
```
